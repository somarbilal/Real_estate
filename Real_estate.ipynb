```python
#Import the moduls
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#import the dataset
df=pd.read_csv("https://raw.githubusercontent.com/Codecademy/datasets/master/streeteasy/manhattan.csv")
```

```python
#Examine the Data
df.head()
```

```python
#Create a DataFrame x that selects somae columns from the main df
x = df[['bedrooms', 'bathrooms', 'size_sqft', 'min_to_subway', 'floor', 'building_age_yrs',
        'no_fee', 'has_roofdeck', 'has_washer_dryer', 'has_doorman', 'has_elevator',
        'has_dishwasher', 'has_patio', 'has_gym']]
```

```python
#Create a DataFrame y that selects the rent column from the main df DataFrame.
y = df[['rent']]
```

```python
#split x into 80% training set and 20% testing set
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2, random_state=6)
```

```python
#look at the shapes of x_train, x_test, y_train, and y_test
print(x_train.shape)
print(x_test.shape)
```

```python
#Create a Linear Regression model and Fit the model using x_train and y_train
mlr = LinearRegression()

mlr.fit(x_train, y_train)
```

```python
#Use the model to predict y-values from x_test
y_predict = mlr.predict(x_test)
```

```python
#predict the rent price of sonny_apartment according to the features we collect
sonny_apartment = [[1, 1, 620, 16, 1, 98, 1, 0, 1, 0, 0, 1, 1, 0]]

predict = mlr.predict(sonny_apartment)

print("Predicted rent: $%.2f" % predict)

print( y_predict)
```

```python
#Create a 2D scatter plot using y_test and y_predict
plt.scatter(y_test, y_predict, alpha=0.4)

# Create x-axis label and y-axis label
plt.xlabel("Prices: $Y_i$")
plt.ylabel("Predicted prices: $\hat{Y}_i$")
plt.title("Actual Rent vs Predicted Rent")

# Show the plot
plt.show()
```

```python
#how to tune and evaluate the model
#Print out the coefficients using .coef_
print(mlr.coef_)
```

```python
#Create a scatterplot of size_sqft and rent
plt.scatter(df[['size_sqft']], df[['rent']], alpha=0.4)

#Create a scatterplot of min_to_subway and rent
plt.scatter(df[['min_to_subway']], df[['rent']], alpha=0.4)

plt.show()
```

```python
#Evaluating the Model's Accuracy using the .score()
# Input code here:
print("Train score:")
print(mlr.score(x_train, y_train))
print("Test score:")
print(mlr.score(x_test, y_test))
```

```python
residuals =y_predict-y_test 
print(residuals)
plt.scatter(y_predict, residuals, alpha=0.4)
plt.title('Residual Analysis')

plt.show()
```
